```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split 

path='RealEstateData.csv'
data=pd.read_csv(path)
data.head(10)
```

```python
data.describe()
```

```python
data.columns
```

# Converting Price (Cr/L) to L (float)

```python
def convert_price(price):
    try:
        price = price.replace('₹', '')  # Remove the Rupee symbol
        if 'Cr' in price:
            return float(price.replace('Cr', '')) * 100  # Convert Cr to lakhs
        elif 'L' in price:
            return float(price.replace('L', ''))  # Keep L as it is
        else:
            return float(price)  # Handle unexpected values
    except ValueError:
        return None

data['Price'] = data['Price'].apply(convert_price)
```

```python
data.head(10)
```

```python
data.describe()
```

```python
data.dropna(axis=0)
```

Creating the Features & Target

```python
data_features=['Total_Area', 'Price_per_SQFT','Baths']
X=data[data_features]
```

```python
y=data.Price
```

Split the dataset into Train & Test

```python
train_X,test_X,train_y, test_y= train_test_split(X,y,random_state=0)
```

Training the Model

```python
data_model= DecisionTreeRegressor(random_state=1)
data_model.fit(train_X,train_y)
```

```python
predict = data_model.predict(test_X)
```

Model Validation using the MAE (Model Ansolute Error)

```python
print(mean_absolute_error(test_y, predict))
```
